## OpenCVを使った画像処理例

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
murata = cv2.imread("murata.jpg")
print(type(murata))# numpy形式で読み込まれる
print(murata.shape)
plt.imshow(murata)
plt.axis("off")
```

cv2.imreadで読み込むとBGRフォーマットなので、  
RGBに変換が必要（BlueとRedが入れ替わってる）  

```python
murata_rgb = cv2.cvtColor(murata, cv2.COLOR_BGR2RGB)
print(murata_rgb.shape)
plt.imshow(murata_rgb)
```

```python
murata_gray = cv2.cvtColor(murata, cv2.COLOR_RGB2GRAY)
print(murata_gray.shape)
plt.imshow(murata_gray, cmap = "gray")
```

グレイ化

```python
plt.imshow(murata_rgb[:-70,160:-120])
print(murata_rgb[:-70,160:-120].shape)
```

拡大

```python
_, threshold_img = cv2.threshold(murata_gray, 100, 200, cv2.THRESH_BINARY_INV)
threshold_img = cv2.cvtColor(threshold_img, cv2.COLOR_GRAY2RGB)
plt.imshow(threshold_img)
```

閾値処理

```python
murata_f = np.fliplr(murata_rgb)
plt.imshow(murata_f)
```

画像フリップ

```python
murata_sobel = cv2.Canny(murata_gray, 150, 200)
plt.imshow(murata_sobel, cmap = "gray")
```
